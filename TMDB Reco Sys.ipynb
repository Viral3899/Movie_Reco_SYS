<b> We are Using Method of Weighted Average of Rating(Votes) for Recommendation
Formula for Weighetd Average</b><br>


                       R*v  +  C*m
            W   =   -----------------
                         v  +   m


<b>W=Weighted Rating</b><br>
<b>R=average for the movies as number from 0 to 10 (mean)=rating</b><br>
<b>v=number of Vote Count</b><br>
<b>m=minimum Votes requried for listing in 70 %quantile of top movies</b><br>
<b>C=the  mean vote across the whole report</b><br>

```python
R=df['vote_average']
v=df['vote_count']
m=df['vote_count'].quantile(0.70)
C=df['vote_average'].mean()
```

```python
df['Weighted average']=((R*v)+(C*m))/(v+m)
```

<b>We Can Also use the Popularity and Weighted Average for Making Decision for recommandation</b><br>

<b>We take Normalized Value of Weigted average and Popularity</b><br>
<b>then we Take 50-50 % Weight from bot the measure and sum it for Recommandatoin</b>

```python
from sklearn.preprocessing import MinMaxScaler

scaler=MinMaxScaler()

Scaled_df=scaler.fit_transform(df[['Weighted average','popularity']])
Scaled_df=pd.DataFrame(Scaled_df,columns=['Weighted average normalized','popularity Normalized'])
```

```python
Scaled_df
```

```python
df[['Weighted average normalized','popularity Normalized']]=Scaled_df
```

```python
df
```

```python
df['score']=df['Weighted average normalized']*0.50+df['popularity Normalized']*0.50
```

```python
top20Popularity=df.sort_values('score',ascending=False).head(20)
```

```python
top20Popularity
```

```python
plt.figure(figsize=(15,10))
sns.barplot(data=top20Popularity,x='score',y='original_title')
```
