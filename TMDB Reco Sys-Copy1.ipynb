### Mergeing dataset

```python
df=pd.merge(movies,credits,how='inner',left_on=movies['id'],right_on=credits['movie_id'])
```

```python
df=df[['id','original_title','original_language','overview','genres', 'keywords','cast','crew']]
```

### Checking Null Values and Droping Non useable Columns

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
```

```python
df
```

```python
df['genres'][0]
```

```python
def conver(obj):
    l=[]
    for i in ast.literal_eval(obj):
        l.append(i['name'])
    return l
```

```python
df['genres']=df['genres'].apply(conver)
```

```python
df['keywords']=df['keywords'].apply(conver)
```

```python
def conver_cst(obj):
    l=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter!=3:
            l.append(i['name'])
            counter+=1
        else:
            break
    return l
```

```python
df['cast']=df['cast'].apply(conver_cst)
```

```python
df.head()
```

```python
def conver_crw(obj):
    l=[]
    counter=0
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            l.append(i['name'])
            break
    return l
```

```python
df['crew']=df['crew'].apply(conver_crw)
```

```python
df['overview']=df['overview'].apply(lambda x:x.split())
```

```python
df.head()
```

```python
df['cast']=df['cast'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
df['crew']=df['crew'].apply(lambda x:[i.replace(" ","") for i in x])
df['keywords']=df['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
df['original_language']=df['original_language'].apply(lambda x:[x])
```

```python
df['genres']=df['genres'].apply(lambda x:[i.replace(" ","") for i in x])
df.head()
```

```python
df['tags']=(df['original_language']+df['overview']+df['genres']+df['keywords']+df['cast']+df['crew'])
```

```python
new_df=df[['id','original_title','tags']]
new_df.head()
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.to_csv('movies.csv')
```

```python
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        ps.stem(i)
    string=" ".join(i)
    y.append(string)
    return y
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append( ps.stem(i))
        
    return " ".join(y)
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
tfidf=TfidfVectorizer(ngram_range=(1,3),max_features=5000,stop_words='english')
```

```python
vctrs=tfidf.fit_transform(new_df['tags']).toarray()
```

```python
vctrs
```

```python
tfidf.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vctrs)
```

```python
pickle.dump(similarity,open('similarity.pkl','wb'))
```

```python
similarity.shape
```

```python
sorted(enumerate(list(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommand(movie):
    movie_index=new_df[new_df['original_title']==movie].index[0]
    dis=similarity[movie_index]
    movie_list=sorted(enumerate(list(dis)),reverse=True,key=lambda x:x[1])[1:6]
    
    for i in movie_list:
        print(new_df.iloc[i[0]]['original_title'])
```
